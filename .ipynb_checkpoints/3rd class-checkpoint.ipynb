```python
import numpy as np
myarray=np.array([
    [[111,112,113],[121,122,123],[131,132,132]],
    [[211,212,213],[221,222,223],[231,232,232]],
    [[311,312,313],[321,322,323],[331,332,332]],
    [[411,412,413],[421,422,423],[431,432,432]]
])
print(myarray)
print("ndim", myarray.ndim, " shape",myarray.shape)

print(myarray[1:3,1,2])
myarray[3,1,2]
# left wale value outer array ke liye rgb value , ke baad vale value inner array ke liye like height of picture and , ke next vale value width of picture
```
